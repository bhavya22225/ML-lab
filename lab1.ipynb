```python
#Question 1.
def sum(a):
    count=0
    for i in range(len(a)):
        for j in range(i+1, len(a)):
            if a[i]+a[j]==10:
                count+=1
    return count

a=[2, 7, 4, 1, 3, 6]

pairs=sum(a)
print(f"Number of pairs with sum 10={pairs}")


#Question 2.
def find_min(a):
    x=a[0]
    for i in a:
        if i<x:
            x=i
    return x

def find_max(a):
    y=a[0]
    for i in a:
        if i>y:
            y=i
    return y
    
a=[5, 3, 8, 1, 0, 4]
if len(a)<3:
    print("range determination not possible")
else:
    mini=find_min(a)
    maxi=find_max(a)
    r=maxi-mini
    print(f"range={r}")


#Question 3.
def matrix_multi(A, B):
    n=len(A)
    r=[[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                r[i][j]+=A[i][k]*B[k][j]
    return r
    

def power(A,m):
    r=A
    for _ in range(m - 1):
        r=matrix_multi(r, A)
    return r


A=[[4, 7, 2], [5, 3, 2], [9, 4, 5]]
m=int(input("enter value of m"))
r=power(A, m)

print("A^m=")
for i in r:
    print(i)
    

#Question 4.
def high(a):
    count={}
    for i in a:
        if i.isalpha():
            if i in count:
                count[i]+=1
            else:
                count[i]=1
                
    max_l=max(count, key=count.get)
    max_count= count[max_l]
    
    return max_l, max_count
                



a=input("enter string")
l, count= high(a)
print(f"Highest occurring character is {l} of count {count}")

                
```
